```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.show()
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple',linewidth=3, alpha=0.5) #RGB Hex Code, basic color okay
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple',lw=3, linestyle=':') # '-',':','step' , ....so on
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple',lw=1, linestyle='-', marker='o', markersize=10,
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color='purple',lw=2, ls='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
