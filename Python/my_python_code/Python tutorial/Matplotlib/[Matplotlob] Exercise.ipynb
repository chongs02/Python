```python
import numpy as np
x = np.arange(0,100)
y=x*2
z=x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Exercise1
# plt.plot(x,y)
# plt.title('Title')
# plt.xlabel('x')
# plt.ylabel('y')


fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
#Exercise2
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y, color='red')


ax2 = fig.add_axes([0.2,.5,.2,.2])
ax2.plot(x,y)
```

```python
#Excercise3
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,z)
#Large
ax.set_xlabel('x')
ax.set_ylabel('z')

#Insert
ax2 = fig.add_axes([.2,.5,.4,.4])
ax2.plot(x,y)
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
```
